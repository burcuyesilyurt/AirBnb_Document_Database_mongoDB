### Question 4

One of employees is thinking of buying a property to rent out. Which bed type is most common in the listings that have waterfront and a dishwasher in New York?

```python
pipeline = [
    # Filtering listings in New York that have a waterfront and a dishwasher
    {
        "$match": {
            "address.market": "New York",  # Filtering New York
            "amenities": {"$regex": ".*Waterfront.*", "$options": "i"},  #searching for the waterfronts
            "amenities": {"$regex": ".*Dishwasher.*", "$options": "i"}  #searching for the dishwashers
        }
    },
    # now we group by bedtype and counting the nmber of listings for each type
    {
        "$group": {
            "_id": "$bed_type",
            "count": {"$sum": 1}
        }
    },
    # sort the number of listings in *decroissant* order 
    {
        "$sort": {"count": -1}
    },
    # limit the result to one document and which means limit to the most common type of bed
    {
        "$limit": 1
    }
]

# executing the pipeline
result = db.listing.aggregate(pipeline)

# printing the result
for doc in result:
    print("The most common bedtype in the listings that has a waterfront and a dishwasher in New York is:", doc["_id"])
```

### Question 5

We are thinking of identifying someone to hire to write review for us professionally. What is the name of the reviewer who left the longest review in New York?


```python
pipeline = [
    # filtering the reviews in New York (only non empty reviews)
    {
        "$match": {
            "address.market": "New York",  # filtering New York
            "reviews.comments": {"$exists": True, "$ne": ""}
        }
    },
    # showing all the reviews
    {
        "$unwind": "$reviews"
    },
    # filtering the reviews to select the only the non-empty reviews
    {
        "$match": {
            "reviews.comments": {"$type": "string"}  # we need to make sure that the comment is a string of characters
        }
    },
    # sorting the reviews in descending order
    {
        "$project": {
            "_id": "$reviews._id",
            "reviewer_name": "$reviews.reviewer_name",
            "review_length": {
                "$cond": {
                    "if": {"$eq": [{"$type": "$reviews.comments"}, "string"]},  # to check if the comment is a string of characters
                    "then": {"$strLenCP": "$reviews.comments"},  # apply the $strLenCP function if not
                    "else": 0  #O otherwise
                }
            }
        }
    },
    {
        "$sort": {"review_length": -1}  #sorting in descending order
    },
    # Limit the result to one review (the longest)
    {
        "$limit": 1
    }
]

# Executing the pipeline
result = db.listingsAndReviews_new.aggregate(pipeline)

# showing the result
for doc in result:
    print("The reviewer who left the longest review in New York is:", doc["reviewer_name"])
```

```python
pipeline = [
    #security_deposit field exists and is not null
    {
        '$match': {
            'security_deposit': {'$exists': True, '$ne': ''}
        }
    },
    #convert relevant fields to numeric values
    {
        '$addFields': {
            'price_': {'$toDouble': {'$ifNull': ['$price', '0']}},  #convert price to double (default to 0)
            'deposit': {'$toDouble': {'$ifNull': ['$security_deposit', '0']}},  #convert deposit to double (default to 0)
            'visitors': {'$ifNull': ['$accommodates', 1]}  #create visitors and set default 1 if accommodates is missing
        }
    },
    #calculate differences per visitor
    {
        '$addFields': {
            'difference': {
                '$cond': {
                    'if': {'$eq': ['$visitors', 0]},  #check if accommodates_numeric is 0 (avoid division by zero)
                    'then': 0,  #if yes set difference to 0 
                    'else': {'$divide': [{ '$abs':{'$subtract': ['$price_', '$deposit']}}, '$visitors']}
                }
            }
        }
    },
    #sort difference in desc order
    {
        '$sort': {'difference': -1}
    },
    #group to find the max and min differences
    {
        '$group': {
            '_id': None,
            'max_difference': {'$first': '$difference'},  #max difference
            'min_difference': {'$last': '$difference'}  #min difference
        }
    },
    #project the results to include relevant fields
    {
        '$project': {
            '_id': 0,
            'max_difference': 1,
            'min_difference': 1
        }
    }
]


cursor = db.listing.aggregate(pipeline)


for result in cursor:
    print("Max difference per visitor:", result['max_difference'])
    print("Min difference per visitor:", result['min_difference'])
```

We want to identify areas by whether they are typically used for short breaks, like weekend mini breaks, or whether they are more suitable for long trips. We will use this information to target advertising of different customers. It is not expected this information will change much over time so we won’t look to update it we just would like a current view. What is the average duration of stay (in nights) per type of property per city (you can use the maximum_nights to measure length of stays)? For each property type return the city with the highest and lowest average value.

```python
pipeline = [
        #maximum_nights field exists and is not null
    {
        "$match": {
            "maximum_nights": {"$exists": True, "$ne": None}
        }
    },
    #convert maximum_nights field to integer
    {
        "$addFields": {
            "max_nights_int": {"$toInt": "$maximum_nights"}
        }
    },
    #group by property_type, city, and calculate average stay
    {
        "$group": {
            "_id": {
                "property_type": "$property_type",
                "city": "$address.market"
            },
            "avgStay": {"$avg": "$max_nights_int"}
        }
    },
    #sort by property_type and averageStay to prepare for finding highest and lowest values
    {
        "$sort": {"_id.property_type": 1, "avgStay": 1}
    },
    #group by property_type to collect city averages and identify most suitable cities
    {
        "$group": {
            "_id": "$_id.property_type",
            "cityAverages": {
                "$push": {
                    "city": "$_id.city",
                    "averageStay": "$avgStay"
                }
            }
        }
    },
    # Project the results to format the output for each property type
    {
        "$project": {
            "_id": 0,
            "propertyType": "$_id",
            "cityWithLowestAvgStay": {"$arrayElemAt": ["$cityAverages", 0]},  # City with lowest average stay
            "cityWithHighestAvgStay": {"$arrayElemAt": ["$cityAverages", -1]}  # City with highest average stay
        }
    }
]

# Execute the aggregation pipeline
result = db.listingsAndReviews_new.aggregate(pipeline)

# Print the results in a user-friendly manner
for document in result:
    property_type = document.get("propertyType", "Unknown")
    short_stay_city = document.get("cityWithLowestAvgStay", {}).get("city", "Unknown")
    short_stay_avg = document.get("cityWithLowestAvgStay", {}).get("averageStay", "Unknown")
    long_stay_city = document.get("cityWithHighestAvgStay", {}).get("city", "Unknown")
    long_stay_avg = document.get("cityWithHighestAvgStay", {}).get("averageStay", "Unknown")
    
    print(f"Property Type: {property_type}")
    print(f"Most Suitable City for Short Stays: {short_stay_city} (Average Stay: {short_stay_avg:.2f} nights)")
    print(f"Most Suitable City for Long Stays: {long_stay_city} (Average Stay: {long_stay_avg:.2f} nights)")
    print("-" * 80)
```
